### Code to reveal the Secret 

```python
import random

input_words = ['apple',  'banana',  'orange',  'coconut',  'strawberry',  'lime',  'grapefruit',  'lemon',  'kumquat',  'blueberry',  'melon']


done = False
secret_word = random.choice(input_words)
right_guesses = []
wrong_guesses = []
guess = ''
faulty_attempt = False
while True:
    while True:
           
        if not wrong_guesses and not right_guesses and not faulty_attempt:
            print('Welcome to Guessing World, You will have 7 lives to guess the secret word, Start your game now.')
        print('\n\n')

        for letter in secret_word:
            if letter in right_guesses:
                print(letter, end='')
            else:
                print('_',end=' ')
                      
        if not right_guesses and not wrong_guesses:
            print('\nYou got a {} letter secret word to guess'.format(len(secret_word)))
            guess = input("\nMake your first guess: ").lower()
        else:
            guess = input("\nMake your next guess to find Secret Word: ").lower()

    
        if len(guess) != 1:
            print("\nYou can only guess a single letter!")
            faulty_attempt = True
        elif guess in right_guesses or guess in wrong_guesses:
            print("\nYou have already guessed that letter!")
            faulty_attempt = True
        elif not guess.isalpha():
            print("\nYou can only guess letters, as the secret words only have letters!")
            faulty_attempt = True
        else:
            break
            
    if guess in secret_word:
        for char_count in range(secret_word.count(guess)):
            right_guesses.append(guess)
        letters_to_guess = len(secret_word)-len(right_guesses)
        if not letters_to_guess:
            found = True
        elif letters_to_guess == 1:
            print('\nCome on! You are left with one last letter to reveal the secret word')
            found = False
        else:
            print('\nThats a good guess. Keep going, you just need {} more letters'.format(letters_to_guess))
            found = False
        if found:
            print("\nHurray, You won by guessing it right. Secret word was {}".format(secret_word))
            break

    else:
        wrong_guesses.append(guess)
        if len(wrong_guesses) == 7:
            print('\nSorry! You lost all your chances. Game Over!')
            print("\nThe secret word was {}".format(secret_word))
            break
        elif len(wrong_guesses) == 6:
            print('Be more Cautious !! you just have one last life to guess the secret word')
        else:
            print('\nSorry! Thats a wrong guess, do not worry you still have {} more lives'.format(7-len(wrong_guesses)))
```
